```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\khans\\Downloads\\iris_dataset.csv")

```

```python
# Dataset structure
df.info()
```

```python
df.describe()
```

```python
grouped_data = df.groupby("species").mean()
grouped_data

```

```python
print("Average measurements of each Iris species:")
grouped_data
```

```python
grouped_data.plot(kind="bar", figsize=(10,6))

plt.title("Average Iris Flower Measurements by Species")
plt.xlabel("Species")
plt.ylabel("Mean Value (cm)")
plt.xticks(rotation=0)
plt.show()
```

```python
### Insights:
print("Insights:")
print("- Iris Virginica has the highest petal length and width.")
print("- Iris Setosa has the smallest petal dimensions.")
print("- Aggregation helps in comparing species easily.")

```
