```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load or create your DataFrame
df = pd.read_csv("C:\\Users\\khans\\OneDrive\\Desktop\\DATASET_INTERNSHIP.csv")
# or however you create df

plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing values before Cleaning")
plt.show()
```

```python
#After cleaning visualizatiob
```

```python
df_cleaned = df.dropna()
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(df_cleaned.isnull(), cbar=False)
plt.title("Missing values after Cleaning")
plt.show()
```

```python
df.isnull().sum()
```

```python
#After cleaning the dataset, missing values have been removed or filled, so df.isnull() returns fewer or no True values.
The heatmap after cleaning therefore shows little to no missing data, confirming that the cleaning step was successful.
```

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\khans\OneDrive\Desktop\DATASET_INTERNSHIP.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

df.select_dtypes(include='number').hist(figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

df.select_dtypes(include='number').plot(kind='box', figsize=(12, 6))
plt.title("Box Plot of Numeric Features")
plt.show()
```

```python
import matplotlib.pyplot as plt

num_cols = df.select_dtypes(include='number').columns

plt.scatter(df[num_cols[0]], df[num_cols[1]])
plt.xlabel(num_cols[0])
plt.ylabel(num_cols[1])
plt.title("Scatter Plot")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.select_dtypes(include='number').corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
